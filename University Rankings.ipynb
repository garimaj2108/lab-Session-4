```python
url = 'http://higheredublog.com/top-universities-for-ms-in-mis-in-usa/'
```

```python
read_data.columns = ['RANKING', 'UNIVERSITY', 'PROGRAM']
```

```python
read_data.dtypes
```

```python
df = pd.DataFrame(read_data)
```
